```python
import cv2
import matplotlib.pyplot as plt
```

```python
def draw_bounding_box(image_path, bbox_str):
    # Parse the bounding box string
    x, y, w, h = map(int, bbox_str.split())

    # Load the image
    image = cv2.imread(image_path)
    
    # Draw the bounding box
    start_point = (x, y)
    end_point = (x + w, y + h)
    color = (0, 255, 0)  # Green color for the bounding box
    thickness = 2  # Thickness of the bounding box line
    
    cv2.rectangle(image, start_point, end_point, color, thickness)
    
    # Convert the image to RGB (since OpenCV uses BGR)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Display the image using matplotlib
    plt.imshow(image_rgb)
    plt.axis('off')  # Hide axes
    plt.show()
    
    cv2.imwrite('output_image.png', image)

image_path = '/interns/iittcseitr24_10/face_attendance/modified_lfw_datasets/embedding_lfw/Amelie_Mauresmo/Amelie_Mauresmo_14.jpg'
bbox_str = "56 5 137 192" 
draw_bounding_box(image_path, bbox_str)
```
