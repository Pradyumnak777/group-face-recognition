```python
faces = DeepFace.extract_faces(img_path='/data/facebiometrics/attendance_app/ML_work/group_image_44.jpg')

# if faces detected
if faces:
    fig, axes = plt.subplots(1, len(faces), figsize=(15, 5))
    if len(faces) == 1:
        axes = [axes]

    # show all faces
    for i, face_obj in enumerate(faces):
        face_img = face_obj["face"]
        axes[i].imshow(face_img)
        axes[i].axis("off")
        axes[i].set_title(f"Face {i + 1}")

    plt.show()
else:
    print("No faces detected")
```
